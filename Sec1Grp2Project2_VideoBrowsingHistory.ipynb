1. Have the user upload their user data json file
2. The script gets the video browsing history:
  - Selenium script that given a set of unique URLs, extracts the video data + comments and stores them in JSON files
3. Print the number of videos in it
4. Saves the list of videos in a generated file following the naming convention "SecXGrY_nnnnn.json"
  - 'nnnnn' are randomly generated numbers
5. Share the file with the user


```python
import os
from google.colab import files
import json
from pandas import Series, DataFrame
import pandas as pd
import dateutil.parser as dparser
from random import randint
```

```python
####### Start of Amy Tasks #########
```

```python
# Select 'user_data.json' file
files.upload()
```

```python
# Read the uploaded user_data.json file
with open('user_data.json', 'r') as myFile:
  data = json.load(myFile)
```

```python
# Extract Video Browsing History and convert to dataframe with date and link
videos_viewed = pd.DataFrame(data['Activity']['Video Browsing History']['VideoList'])
videos_viewed.head()
```

```python
# Number of Videos
videos_viewed.shape[0]
```

```python
# Generate New File Name
randNum = randint(10**(5-1), (10**5)-1)
str(randNum)
filename = f"Sec2Gr2_{randNum}.json"
filename
```

```python
# Save to json file
videos_viewed.to_json(filename, orient = 'split', compression = 'infer', index = 'true')
```
